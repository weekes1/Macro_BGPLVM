```python
import os
script_directory = os.path.dirname(os.path.abspath("Macro_Data_Combinor.ipynb"))
print(script_directory)


Folder = "CSV_Files" # make sure folder is named as such. !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

Path_to_Macro_data = os.path.join(script_directory,Folder)
print(Path_to_Macro_data)

Macro_data_files_names = os.listdir(Path_to_Macro_data)
print("File num: ", len(Macro_data_files_names))
print("File names: ", Macro_data_files_names)

file_paths_Macro_data = [os.path.join(Path_to_Macro_data,file) for file in Macro_data_files_names] #join base of path to each file name
print(file_paths_Macro_data)


#
```

```python
Macro_df_all_available_dates = pd.read_csv(file_paths_Macro_data[0])

#print(Macro_df)

for i,file in enumerate(file_paths_Macro_data[1:]):
    #read in excell file
    df = pd.read_csv(file)
    file_name = Macro_data_files_names[i+1]
    var_name = file_name[0:-4]
    df.columns.values[1] = var_name
    Macro_df_all_available_dates = pd.merge(Macro_df_all_available_dates, df, on='DATE', how='outer')
    
Macro_df_all_available_dates.set_index('DATE',inplace=True)
print(Macro_df_all_available_dates.shape)
print(Macro_df_all_available_dates)   

Macro_df_all_available_dates.to_csv('Master_Macro_Data_Set.csv')
```

# Select for dates from 1971 to 2023

```python
Macro_Data_71_23 = Macro_df_all_available_dates.loc['1971-01-01':'2023-01-01']
print(Macro_Data_71_23.columns)
#print(Macro_Data_71_23)

Macro_Data_71_23.to_csv("Macro_Data_71_23.csv")

Macro_Data_71_23_micro = Macro_Data_71_23[['GDP',"IR%","CO2",'unemp%','Inf%','GINI_INDEX','C_Sent(IND)','LifeExp']]

Macro_Data_71_23_micro = Macro_Data_71_23_micro.ffill()
print(Macro_Data_71_23_micro)

Macro_Data_71_23_micro.to_csv("Macro_Data_71_23_small.csv")
```

# Calculate Percent changes

```python
Macro_Data_71_23 = Macro_Data_71_23.ffill()
Macro_Data_71_23 = Macro_Data_71_23.astype(float)


Macro_Data_71_23_pct = Macro_Data_71_23.pct_change() * 100
Macro_Data_71_23_pct = Macro_Data_71_23_pct.iloc[1:]
print(Macro_Data_71_23_pct)

Macro_Data_71_23_pct.to_csv("Macro_Data_71_23_pct.csv")
```
